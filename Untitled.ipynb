```python
# === Configura aqu ===
RUTA = "./tercera_entrega"   # tu carpeta raz
PATRON = "*"                  # "*" = TODOS los archivos, sin filtrar por extensin
RECURSIVO = True              # True = incluye subcarpetas

from pathlib import Path

base = Path(RUTA)
if not base.is_dir():
    print(f"Error: '{base}' no es una carpeta.")
else:
    paths = base.rglob(PATRON) if RECURSIVO else base.glob(PATRON)
    encontrados = False

    for p in sorted(paths):
        if not p.is_file():
            continue
        encontrados = True

        # Intento de lectura como texto probando codificaciones comunes
        contenido = None
        encoding_usada = None
        for enc in ("utf-8", "utf-8-sig", "latin-1", "cp1252"):
            try:
                with open(p, "r", encoding=enc, errors="strict") as f:
                    contenido = f.read()
                encoding_usada = enc
                break
            except UnicodeDecodeError:
                continue

        # Fallback: imprime reemplazando caracteres invlidos
        if contenido is None:
            with open(p, "r", encoding="utf-8", errors="replace") as f:
                contenido = f.read()
            encoding_usada = "utf-8 (replace)"

        print(f"\n===== {p} (encoding: {encoding_usada}) =====")
        print(contenido, end="" if contenido.endswith("\n") else "\n")
        print(f"===== FIN de {p} =====")

    if not encontrados:
        print("No se encontraron archivos que coincidan con el patrn.")
```
